```
#El siguiente código sirve para resolver ecuaciones cuadráticas con la fórmula general.

import cmath #Librería necesaria para el cálculo de las raíces complejas

#Raíces positivas
def pos(x,y,z):
  return (-y+cmath.sqrt(z))/(2*x)

#Raíces negativas
def neg(x,y,z):
  return (-y-cmath.sqrt(z))/(2*x)
  
print("Este programa sirve para encontrar las soluciones de una ecuación cudrática") #Dándole a conocer la función del programa al usuario
print("(de la forma ax^2+bx+c=0) haciendo uso de la fórmula general.")

print("\nPor favor introduce el valor de a de la ecuación que deseas resolver.") #Solicitando datos
a=float(input())
print("\nAhora, por favor introduce el valor de b.")
b=float(input())
print("\nAhora, por favor introduce el valor de c.")
c=float(input())

f=b**2-(4*a*c)

p=pos(a,b,f) #Raíz positiva
n=neg(a,b,f) #Raíz negativa

print("\nLa raíz x1 de la ecuación especificada es:", p)
print("\nLa raiz x2 de la ecuación especificada es:", n)
```

**Ejercicio 3: Método de Herón**


$x_{n+1}=\frac{1}{2}(x_n+\frac{S}{x_n})$, donde $x_{n+1}$ es la media entre $S$ y $x_n$

```
# El siguiente código sirve para aproximar la raíz de un número arbitrario con el método de Herón.

import numpy as np

print("Este programa sirve para aproximar la raíz de un número arbitrario con el método de Herón.") #Dándole a conocer la función del código al usuario
print("\nPor favor introduzca el número cuya raíz desea aproximar.")

S=float(input())
Srai=np.sqrt(S) #Esta es la raíz "exacta" del número

print("\nAhora, por favor introduce un valor inicial x0.")

x0=float(input())

i=1 #Inicio del procedimiento
aprox=1/2*(x0+(S/x0))

while abs(S-aprox) > 0.01: #Desarrollo del ciclo
  print("Ciclo", i, ":", aprox)
  aprox=1/2*(aprox+(S/aprox))
  i=i+1
  if i >= 10000: 
    break

print("La aproximación de la raíz de ", S, "es:", aprox)
print("\nEl eror de la aproximación es de:", abs(np.sqrt(S)-aprox))
```

**Ejercicio 4: Serie de Taylor**

$ln(1+x)=\displaystyle\sum_{n=1}^\infty (-1)^{n+1}\frac{x^n}{n}=x-\frac{x^2}{2}+\frac{3^3}{3}$

```
# Este código usa la serie de Taylor para aproximar una función especifica: ln(1+x).

import numpy as np #Importando librerías
import matplotlib.pyplot as plt 

print("Este programa emplea la serie de Taylor para aproximar la función ln(1+x).") #Dándole a conocer la función del programa al usuario
print("\nPor favor introduzca el grado del polinomio que desea emplear.") #Solicitando datos
print("NOTA: Recuerde que debe ser un número natural.")

n=int(input())

cero=0.0

if (n < 0): #Esta línea evita que el usuario ignore la nota impresa anteriormente
  print("Por favor introduzca un número positivo.")
  quit()

x=np.linspace(-1.5,1.5,100)

Tnau=0.0 #Valor auxiliar
for i in range(1,n+1,1):
  Tn=Tnau+((-1)**(i+1))*(x**i/i)
  Tnau=Tn

print("\nAhora, por favor introduce un grado diferente en el que quieras probar.") # Dándole la opción al usuario de que aproxime el
                                                                                   # valor con un polinomio de diferente grado

n1=int(input())
Tnau1=0.0

if (n1 < 0):
  print("Por favor introduzca un número positivo.")
  quit()

for i in range(1,n1+1,1):
  Tn1=Tnau1+((-1)**(i+1))*(x**i/i)
  Tnau1=Tn1

print("\nIntroduce otro grado.")

n2=int(input())
Tnau2=0.0 
if (n2 < 0):
  print("Por favor introduzca un número positivo.")
  quit()
  
for i in range(1,n2+1,1):
  Tn2=Tnau2+((-1)**(i+1))*(x**i/i)
  Tnau2=Tn2

print("\nUno más.")

n3=int(input())
Tnau3=0.0
if (n3 < 0):
  print("Por favor introduzca un número positivo.")
  quit()
  
for i in range(1,n3+1,1):
  Tn3=Tnau3+((-1)**(i+1))*(x**i/i)
  Tnau3=Tn3

# Graficación 
plt.subplot(111)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Aprox. de ln(1+x)")


plt.plot(x,Tn, label="$T_4$")
plt.plot(x,Tn1, label="$T_7$")
plt.plot(x,Tn2, label="$T_{11}$")
plt.plot(x,Tn3, label="$T_{16}$")
plt.plot(x,np.log(1+x), label="$ln(1+x)$")
# Malla
plt.grid(True)
# Límites en el eje Y
plt.ylim(-4, 2)

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()

plt.savefig('Actividad2Ejercicio4.png')  
print("Fin del programa.")
```
